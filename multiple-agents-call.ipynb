```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

```python
load_dotenv(override=True)
```

```python
openai_api_key = os.getenv("OPENAI_API_KEY")
gemeni_api_key = os.getenv("GEMENI_API_KEY")
groq_api_key = os.getenv("GROQ_API_KEY")
deepseek_api_key = os.getenv("DEEPSEEK_API_KEY")
openai = OpenAI(api_key=openai_api_key)
```

```python
system_prompt = """You are an expert CV generator using LaTeX. Create a clean, ATS-friendly CV in black and white (no colored elements) for a fresh graduate computer science student named Ahmed Saber from Minya University, Egypt. Follow these guidelines:

1. Overall Formatting
   - Use plain LaTeX (article or a simple CV class is fine), avoid any color or styling packages beyond standard fonts and spacing.
   - Keep the layout simple and well-organized. Use bold section headers and bullet lists where appropriate.
   - Ensure the document compiles without errors.

2. Section Order
   1. Contact Information – include name, location (Minya, Egypt), email, and GitHub link.
   2. Professional Summary – a concise paragraph (2–3 sentences) summarizing Ahmed’s strengths (e.g., strong programming foundations, experience with compilers, web back‑end and front‑end development; passion for problem solving; solid competitive programming record).
   3. Education – B.Sc. in Computer Science from Minya University, with graduation date (assume 2025 if unknown).
   4. Technical Skills – split into subsections:
      - Programming Languages: C/C++, Go (Golang), JavaScript, Python.
      - Web Frameworks & Tools: Fiber (Go web framework), Gin, Node.js, REST API development, HTML/CSS, simple JavaScript applications.
      - Computer Graphics & Compilers: basics of rendering, finite‑state machines, compiler design (lexical/syntax analysis, parsing).
      - DevOps: CI/CD pipelines, Git/GitHub.
   5. Projects – list key projects with succinct descriptions and quantifiable outcomes where possible:
      - C‑Compiler – built a basic compiler demonstrating lexical analysis, parsing, semantic analysis, and code generation.
      - GO‑RestAPI & Fiber Projects – implemented RESTful APIs in Go using Fiber/Gin; leveraged goroutines and context for concurrency.
      - ChatApplication – developed a socket‑based chat server and client showcasing TCP networking and multithreading.
      - AI‑Agents – experimented with multi‑agent systems (autonomy, concurrency, communication).
      - DevOps Pipelines – automated testing/deployment with CI/CD.
      - Food‑Order‑Website & Frontend Mentor Challenges – created responsive HTML/CSS interfaces (blog preview, result summary) and a food ordering platform with cart and checkout logic.
      - Fawrry‑Internship‑Task – integrated FawryPay Express Checkout API, handling card payments, tokenization, and POS options.
      - Finite‑State‑Machine & Graphics Projects – implemented finite‑state machines and basic computer graphics demos.
   6. Competitive Programming & Achievements
      - Highlight Codeforces Expert rating with a peak of 1642; over 3,000 problems solved on Codeforces.
      - Mention solving 300+ problems on AtCoder and 200+ on LeetCode.
      - Note participation in ECPC and IEEE Xtreme contests.
      - Include roles as Problem Solving Mentor at the ICPC Minya Community and Problem Setter for ECPC Qualifications.
   7. Mentoring & Leadership – briefly describe mentoring duties (helping peers with algorithms and problem solving) and teamwork experience during university projects.

3. Writing Style
   - Use clear, concise language optimized for Applicant Tracking Systems (ATS). Avoid long paragraphs; prefer bullet points.
   - Emphasize measurable achievements (e.g., number of problems solved, rating achieved, size/scale of projects).
   - Keep descriptions professional and focused on outcomes and responsibilities.

Provide the complete LaTeX code for the CV, ready to compile.
"""
```

```python
messages=[{"role": "user" , "content": system_prompt}]
mapResponse = []
```

```python
# OpenAI
response = openai.chat.completions.create(
    model="gpt-5-nano",
    messages=messages,
)
mapResponse.append({"model": "gpt-5-nano", "response": response.choices[0].message.content})
print(response.choices[0].message.content)
```

```python
# Gemini
gemini_client = OpenAI(api_key=gemeni_api_key, base_url="https://generativelanguage.googleapis.com/v1beta/openai/")
response = gemini_client.chat.completions.create(
    model="gemini-pro",  # Changed from gemini-2.5-flash (which doesn't exist)
    messages=messages,
)
mapResponse.append({"model": "gemini-pro", "response": response.choices[0].message.content})
print(response.choices[0].message.content)
```

```python
# Groq
groq_client = OpenAI(api_key=groq_api_key, base_url="https://api.grok.com/v1")
response = groq_client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=messages,
)
mapResponse.append({"model": "openai/gpt-oss-120b", "response": response.choices[0].message.content})
print(response.choices[0].message.content)
```

```python
# Deepseek
response = OpenAI(api_key=deepseek_api_key , base_url="https://api.deepseek.com/v1")
response = openai.chat.completions.create(
    model="deepseek-chat",
    messages=messages,
)
mapResponse.append({"model": "deepseek-chat", "response": response.choices[0].message.content})     
print(response.choices[0].message.content)
```
